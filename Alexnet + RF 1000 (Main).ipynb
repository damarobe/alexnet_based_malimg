# This is where Alexnet + RF is tried on the 25000 images

## this makes use os the file generated by the Alexnet feature extraction code

## This assumes the existance of a dir 'alexnet_fearutes_1000' which contains the features for the images

```python
from torchvision import models
import torch
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
from torchvision import transforms
import os
import pandas as pd
import torch.nn as nn
import torch.optim as optim
import time
from torch.utils.data import DataLoader,Dataset
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
dataset= './alexnet_features_1000/'
```

# Read data from the file

```python
x=[]
y=[]
for (root,dirs,files) in  os.walk(dataset, topdown=True):
    dirname = root.split(os.path.sep)[-1]
    
    if len(files)==0:
        continue
    for i in files:
        if i[-3:] == 'png':
            continue
        tensors=torch.load(root+'/'+i) # laod data ... tensor data
#         print(len(tensors))
        for j in tensors:
            x.append(j)
            y.append(dirname)
        
        
```

```python
# encoding ...
```

```python

le = preprocessing.LabelEncoder()
le.fit(y)
labels_encoded = le.transform(y)
y_train = labels_encoded
x_np = [i.detach().numpy()[0] for i in x]
```

# Model setup ... 

```python
RF_model = RandomForestClassifier(n_estimators = 1, random_state = 42)
X_train, X_test, Y_train, Y_test = train_test_split(x_np, y_train, test_size=0.2)
# Train the model on training data
RF_model.fit(X_train, Y_train)
```

```python
pred = RF_model.predict(X_test)
print ("Accuracy = ", metrics.accuracy_score(Y_test, pred))
```
